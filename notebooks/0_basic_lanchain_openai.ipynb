### langchain model openai

load env variable

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-5-mini-2025-08-07")
```

```python
res = llm.invoke("Rajasthan ka CM kon hai?")
```

```python
res.content
```

```python
## connect with googke gemini model
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
res_gemini = llm_gemini.invoke("Rajasthan ka CM kon hai?")
```

```python
res_gemini.content
```

```python
prompts = [("system",'You are a data engineer'), ## system, ai ,user
              ("user","Explain AI data engineer")
]
```
